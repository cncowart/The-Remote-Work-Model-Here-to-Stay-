```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd
```

```python
website_url = 'https://www.bls.gov/opub/ted/2020/unemployment-rate-falls-to-6-point-9-percent-in-october-2020.htm#tab-2'
response = requests.get(website_url)

response.status_code
```

```python
print(type(response))
response.content
```

```python
soup = BeautifulSoup(response.content, 'lxml')
print(soup.title)
```

```python
tables = soup.find_all('table', attrs = {'class': 'regular'})
tables
```

```python
len(tables)
```

```python
result_list = pd.read_html(str(tables[0])) # a list of dataframes
len(result_list)
```

```python
labor_stat = result_list[0]  # get the first df from the list
labor_stat
```

```python
labor_stat.to_csv(r'labor_stat.csv')
```
